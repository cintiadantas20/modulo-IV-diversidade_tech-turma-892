````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
````

#### Questão 1

Criar 2 vetores, um com uma série historica de anos de 1970 até 2022 e um segundo vetor com a mesma quantidade de observações e números inteiros aleatórios variando de 1 a 20000. Em seguida montar um gráfico com a melhor visualização para os dados com a biblioteca Plotly Express.

````python
# Cria o vetor de números aleatórios
np.random.seed(42)
number = np.random.randint(20001, size=53)

# Cria o vetor da série temporal
year = pd.date_range('1970', periods=53, freq='y')

# Cria o dataframe com os dois vetores
df1 = pd.DataFrame()
df1['year'] = pd.Series(year)
df1['number'] = pd.Series(number)
display(df1)
````

````python
# Plota um gráfico de linha com a lib plotly do dataframe criado
x_line = df1['year'].values
y_line = df1['number'].values
fig1 = px.line(x=x_line, y=y_line, markers=True)

# Títulos
fig1.update_layout(
    xaxis_title={
        'text': 'Anos',
        'font_size': 12
    },
    yaxis_title={
        'text': 'Números inteiros aleatórios',
        'font_size': 12
    },
    font_color='grey',
    font_size=10,
    title={
        'text': 'Série temporal de números aleatórios',
        'font_family': 'Arial',
        'font_color': 'black',
        'font_size': 16
    },
    template='ggplot2'
)
````

#### Questão 2

Criar uma lista com as letras do alfabeto, depois um vetor com a mesma quantidade de observações e números inteiros aleatórios variando de 10 a 100. Em seguida montar um gráfico com a melhor visualização para os dados com a biblioteca Plotly Express.

Utilize o objeto string para resolver a questão:

```
import string
string.ascii_lowercase
```

````python
# Cria uma lista com as letras do alfabeto
import string
letras = list(string.ascii_lowercase)

# Cria um array com 26 números aleatórios
np.random.seed(42)
number2 = np.random.randint(10, 101, size=len(letras))

# Cria o dataframe com os dois vetores
df2 = pd.DataFrame()
df2['letras'] = letras
df2['números'] = number2
display(df2)
````

````python
# Plota um gráfico de barras com a lib plotly do dataframe criado
fig2 = px.histogram(data_frame=df2,
             x='letras',
             y='números',
             barmode='group',
             text_auto=True)

# Títulos
fig2.update_layout(
    xaxis_title={
        'text': 'Letras',
        'font_size': 12,
    },
    yaxis_title={
        'text': 'Números inteiros aleatórios',
        'font_size': 12
    },
    font_color='white',
    font_size=10,
    title={
        'text': 'Gráfico de barras de números aleatórios',
        'font_family': 'Arial',
        'font_color': 'white',
        'font_size': 16
    },
    template='plotly_dark'
)
````

#### Questão 3

Criar 2 vetores com a mesma quantidade de observações, 500 registros , de números inteiros aleatórios, o primeiro variando de 1000 a 2000, o segundo variando de 5000 a 10000. Em seguida montar um gráfico com a melhor visualização para os dados com a biblioteca Plotly Express, demonstrando relação entre as variáveis.

````python
# Cria o primeiro vetor de números aleatórios
np.random.seed(42)
vetor1 = np.random.randint(1000, 2001, 501)

# Cria o segundo vetor de números aleatórios
np.random.seed(42)
vetor2 = np.random.randint(5000, 10001, 501)

# Cria o dataframe com os dois vetores
df3 = pd.DataFrame()
df3['vetor1'] = pd.Series(vetor1)
df3['vetor2'] = pd.Series(vetor2)
display(df3)
````

````python
# Plota um gráfico de dispersão com a lib plotly do dataframe criado
fig3 = px.scatter(data_frame=df3, 
           x='vetor1', 
           y='vetor2',
           trendline='ols')

# Títulos
fig3.update_layout(
    xaxis_title={
        'text': 'Vetor 1',
        'font_size': 12,
    },
    yaxis_title={
        'text': 'Vetor 2',
        'font_size': 12
    },
    font_color='grey',
    font_size=10,
    title={
        'text': 'Gráfico de dispersão de números aleatórios',
        'font_family': 'Arial',
        'font_color': 'black',
        'font_size': 16
    },
    template='gridon'
)
````

#### Questão 4

Criar uma vetor com 500 registros de números inteiros variando entre 125 e 857. Em seguida montar um gráfico que mostre a dispersão com a melhor visualização para os dados com a biblioteca Plotly Express.

````python
# Cria o vetor de números aleatórios
np.random.seed(42)
vetor = np.random.randint(125, 858, 501)

# Cria o dataframe com o vetor
df4 = pd.DataFrame()
df4['vetor'] = pd.Series(vetor)
display(df4)
````

````python
# Plota um gráfico de dispersão com a lib plotly do dataframe criado
fig4 = px.box(data_frame=df4, 
           y='vetor')

# Títulos
fig4.update_layout(
    xaxis_title={
        'text': 'Índice',
        'font_size': 12,
    },
    yaxis_title={
        'text': 'Números inteiros aleatórios',
        'font_size': 12
    },
    font_color='grey',
    font_size=10,
    title={
        'text': 'Gráfico de dispersão de números aleatórios',
        'font_family': 'Arial',
        'font_color': 'black',
        'font_size': 16
    },
    template='simple_white'
)
````
